```python
items = []
for page in range(1, 51):
    url = f'https://books.toscrape.com/catalogue/page-{page}.html'

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    product_price = soup.select('article.product_pod')

    for article in product_price:
        title = article.find('h3').find('a')['title']
        price= article.select_one('p.price_color').text.split('£')[1]
        rating = article.select_one('p.star-rating')
        Rating = rating['class'][1]
        items.append([title, price, Rating])
```

```python
import pandas as pd
```

```python
df =  pd.DataFrame(items, columns=['Book Title', 'Price', 'Rating'])
```
