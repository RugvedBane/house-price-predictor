```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
rohit = [0, 300, 600, 1000, 1300, 1800, 1400, 1500, 1900, 0]
```

```python
plt.plot(years, kohli, color="red", linestyle="--", label="Kohli")
plt.plot(years, sehwag, color="green", linestyle="-.", linewidth=2, label="sehwag")
plt.plot(years, rohit, color="Blue", label="Rohit")
plt.legend()
plt.grid()
plt.tight_layout()
```

```python
plt.plot(years, kohli, color="red", linestyle="--", label="Kohli")
plt.plot(years, sehwag, color="green", linestyle="-.", linewidth=2, label="sehwag")
plt.plot(years, rohit, color="Blue", label="Rohit")
plt.legend()
plt.grid()
plt.tight_layout()
plt.style.use("default")
```

```python
plt.bar(years, kohli)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Runs Scored By Kohli")
```

```python
import numpy as np

x = np.arange(len(years))
width = 0.25
```

```python
plt.bar(x- width, kohli, width=width, label="Kohli")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, rohit, width=width, label="Rohit")

plt.xlabel("year")
plt.ylabel("Runs Scored")
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="skyblue")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center')

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
plt.barh(players, runs, color="skyblue")

for i in range(len(players)): 
    plt.text(runs[i] + 20, i, str(runs[i]), va="center")

plt.margins(x=0.1)

plt.xlabel("Runs")
plt.title("Runs Scored By Players")
plt.tight_layout()
```

```python
plt.figure(figsize=(6,6))
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels, wedgeprops={'edgecolor': 'black', 'linewidth' : 1, 'linestyle': '--'})

plt.show()
```

```python
explode=[0.06, 0, 0, 0]

plt.pie(runs, labels=labels, explode=explode, shadow=True, startangle=90, autopct="%1.1f%%")
plt.show()
```
