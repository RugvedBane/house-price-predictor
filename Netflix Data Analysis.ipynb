```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
top_countries = df2[df2['country'] != "Unknown"]['country'].value_counts().head(10)
top_countries.plot(kind="bar")

for i, v in enumerate(top_countries):
    plt.text(i, v + 50, str(v),fontsize = 10, ha="center")

plt.margins(y=0.1)

plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.title("top 10 countries producing netflix content")
```

```python
plt.figure(figsize=(6,6))
type_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
released_year = df2['release_year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
released_year.plot()


plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.title("Netflix Content Released Over Years")
plt.grid()
```

```python
df2['rating'] = df2['rating'].fillna("Unknown")
```

```python
df2.isnull().sum()
```

```python
plt.figure(figsize=(12,6))
rating_counts = df2[df2['rating'] != "Unknown"]['rating'].value_counts()
plt.barh(rating_counts.index, rating_counts.values)

plt.gca().invert_yaxis()

for i, v in enumerate(rating_counts):
    plt.text(v + 20, i, str(v))

plt.xlabel("Number of Titles")
plt.ylabel("count")
plt.title("Distribution of Ratings on netflix")
```

```python
plt.figure(figsize=(12,6))
genres = df2['listed_in'].str.split(', ').explode()

top_genres = genres.value_counts().head(10)

top_genres.plot(kind='barh')

for i, v in enumerate(top_genres):
    plt.text(v + 20, i, str(v), va="center")

plt.gca().invert_yaxis()
plt.title('top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genres')
```

```python
plt.figure(figsize=(6,4))
type_counts = df2['type'].value_counts()
plt.bar(type_counts.index, type_counts.values)

for i, v in enumerate(type_counts):
    plt.text(i, v + 50, str(v), ha="center")

plt.ylabel("Count")
plt.title("Movies vs TV Shows on Netflix")
```

```python
movies = df2[df2['type']== 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace("min", "")
movies['duration'] = pd.to_numeric(movies['duration'], errors="coerce")

plt.figure(figsize=(12,6))
movies['duration'].plot(kind="hist", bins=20)

movies_counts = movies['duration'].value_counts()



plt.xlabel('Movie duration (Minutes)')
plt.ylabel('Number of movies')
plt.title('Distribution of Movie durations on Netflix')
plt.show()

```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
content_per_year = df['year_added'].value_counts().sort_index()
plt.figure(figsize=(10,6))

content_per_year.plot(kind='line', marker='o')

plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.title("Number of Titles Added to Netflix Each Year")

plt.show()
```
