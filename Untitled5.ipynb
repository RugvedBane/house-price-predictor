```python
import json

def cleaned_data(data):
    data['users'] = [user for user in data['users'] if user.get('name') and user['name'].strip() ]

    for user in data['users']:
        if user.get('friends') is None:
            user['friends'] = []
        user['friends']= list(set(f for f in user['friends'] if f is not None))

        if user.get('liked_pages') is None:
            user['liked_pages'] = []
        user['liked_pages'] = list(set(user['liked_pages']))
        
        
        
    
    return data


data = json.load(open("data3.json"))
data = cleaned_data(data)
json.dump(data, open("cleaned_data2.json", "w"), indent = 3)
print("data has been cleaned successfully")
```
