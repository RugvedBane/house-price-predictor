```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import time

# ==============================
# TELEGRAM CONFIG
# ==============================
BOT_TOKEN = "YOUR_BOT_TOKEN"
CHAT_ID = "YOUR_CHAT_ID"

def send_telegram(msg):
    url = f"https://api.telegram.org/bot{BOT_TOKEN}/sendMessage"
    data = {"chat_id": CHAT_ID, "text": msg}
    requests.post(url, data=data)


# ==============================
# PRODUCT LIST
# ==============================
products = [
    {
        "name": "t-shirt",
        "url": "https://www.amazon.in/dp/B0F5GX46NX"
    }
]


# ==============================
# SCRAPER FUNCTION
# ==============================
def get_price(url):
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
    "Accept-Language": "en-IN,en;q=0.9",
    "Referer": "https://www.amazon.in/",
    "Connection": "keep-alive"
}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    try:
        price = soup.find("span", {"class": "a-offscreen"})
        if price is None:
            price = soup.find("span", {"class": "a-price-whole"})

        price = price.text.replace("₹", "").replace(",", "").strip()
        return float(price)

    except:
        return None


# ==============================
# LOAD OLD DATA (SAFE)
# ==============================
FILE = "prices.csv"

if os.path.exists(FILE):
    try:
        old_data = pd.read_csv(FILE)

        if old_data.empty:
            old_data = pd.DataFrame(columns=["name", "price"])

    except pd.errors.EmptyDataError:
        old_data = pd.DataFrame(columns=["name", "price"])
else:
    old_data = pd.DataFrame(columns=["name", "price"])


# ==============================
# MAIN LOGIC
# ==============================
new_data = []

for product in products:
    name = product["name"]
    url = product["url"]

    print(f"Checking {name}...")
    new_price = get_price(url)

    if new_price is None:
        print(f"❌ Failed to get price for {name}")
        continue

    print(f"✅ Current Price: {new_price}")

    # Find old price
    old_row = old_data[old_data["name"] == name]

    if not old_row.empty:
        old_price = float(old_row["price"].values[0])

        if new_price < old_price:
            msg = f"🔥 Price Dropped!\n{name}\n{old_price} → {new_price}"
            send_telegram(msg)
            print(msg)
        else:
            print(f"No drop: {old_price} → {new_price}")
    else:
        print(f"New product added: {name}")

    new_data.append({"name": name, "price": new_price})

    time.sleep(2)  # avoid blocking


# ==============================
# SAVE NEW DATA
# ==============================
if len(new_data) > 0:
    df = pd.DataFrame(new_data)
    df.to_csv(FILE, index=False)
    print("✅ Prices updated in CSV")
else:
    print("⚠️ No data scraped, CSV not updated")


print("🎯 Done!")
```
