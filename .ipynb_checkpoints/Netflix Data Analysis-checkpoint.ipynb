```python
df2 = df.copy()
```

```python
df2
```

```python
df2['director'] = df2['director'].fillna("Unknown")
```

```python
df2
```

```python
df2['cast'] = df2['cast'].fillna("Unknown")
```

```python
df2
```

```python
df2 = df2.dropna(subset=['date_added'])
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
df2['country'] = df2['country'].fillna("Unknown")
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
type_counts = df['type'].value_counts()
```

```python
type_counts
```

```python
import matplotlib.pyplot as plt
```

```python
top_countries = df2[df2['country'] != "Unknown"]['country'].value_counts().head(10)
top_countries.plot(kind="bar")

for i, v in enumerate(top_countries):
    plt.text(i, v + 50, str(v),fontsize = 10, ha="center")

plt.margins(y=0.1)

plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.title("top 10 countries producing netflix content")
plt.tight_layout()
```

```python
type_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
released_year = df2['release_year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
released_year.plot()


plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.title("Netflix Content Released Over Years")
plt.tight_layout()
plt.grid()
```

```python
df2['rating'] = df2['rating'].fillna("Unknown")
```

```python
df2.isnull().sum()
```

```python
plt.figure(figsize=(12,6))
rating_counts = df2[df2['rating'] != "Unknown"]['rating'].value_counts()
plt.barh(rating_counts.index, rating_counts.values)

plt.gca().invert_yaxis()

for i, v in enumerate(rating_counts):
    plt.text(v + 20, i, str(v))

plt.xlabel("rating")
plt.ylabel("count")
plt.title("Distribution of Ratings on netflix")
plt.tight_layout()
```

```python
plt.figure(figsize=(12,6))
df2['rating'].value_counts().plot(kind="barh")
plt.gca().invert_yaxis()


plt.xlabel("rating")
plt.ylabel("count")
plt.title("Distribution of Ratings on netflix")
plt.tight_layout()
```
