```python
import seaborn as sns
import numpy as np
```

```python
sns.set_theme(style='whitegrid')
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data
```

```python
sns.relplot(x='total_bill', y='tip', data=tips_data, hue='sex', size='size', col='day',col_wrap=2, style='smoker', kind='scatter')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_data, hue='day')
```

```python
sns.displot(tips_data, x='tip', hue='sex', kind='hist', alpha=0.6)
sns.rugplot(tips_data, x='tip', color='orange')
```

```python
sns.barplot(tips_data, x='day', y='tip', hue='sex')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.scatterplot(x="age", y="salary", hue="gender", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.show()
```
