```python
import pandas as pd
```

```python
df = pd.read_csv("data cleanning.csv")
```

```python
df
```

```python
df.info()
```

```python
df['Age']= pd.to_numeric(df['Age'], errors="coerce")
```

```python
df['Age'] = df['Age'].ffill()
```

```python
df
```

```python
df['Gender'] = df['Gender'].replace({"m" : "M", "f" : "F"})
```

```python
df
```

```python
df['Marks'] = pd.to_numeric(df['Marks'], errors= "coerce")
```

```python
df
```

```python
df.loc[(df['Marks'] > 100) | (df['Marks'] < 0), "Marks"] = pd.NA
```

```python
df
```

```python
df['Attendance'] = df['Attendance'].str.replace("%", "")
```

```python
df['Attendance'] = pd.to_numeric(df['Attendance'])
```

```python
df.loc[(df['Attendance'] > 100) | (df['Attendance'] < 0), "Attendance"] = pd.NA
```

```python
df
```

```python
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].median())
```

```python
df
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors="coerce", format="mixed")
```

```python
df
```

```python
df['Course'] = df['Course'].str.strip().str.lower()
```

```python
df
```

```python
df['Course'] = df['Course'].str.title()
```

```python
df
```

```python
df['Course'] = df['Course'].replace("Datascience", "Data Science")
```

```python
df
```

```python
df['Name'] = df['Name'].str.title()
```

```python
df
```

```python
df.duplicated(subset=(["Name", "Age", "Marks", "Attendance", "Join_Date", "Course"]))
```

```python
df = df.drop_duplicates(subset=(["Name", "Age", "Marks", "Attendance", "Join_Date", "Course"]))
```

```python
df
```

```python
df = df.drop
df = df.reset_index(drop=True)
```

```python
df
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors="coerce", dayfirst=True)
```

```python
df
```

```python
df = df.drop(columns=["index"])
```

```python
df
```

```python
df['Join_Date'] = df['Join_Date'].dt.strftime("%Y-%m-%d")
```
